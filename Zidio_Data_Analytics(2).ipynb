```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_style("whitegrid")
```

```python
df_players = pd.read_excel('/content/Players.xlsx')
```

```python
df_players.head()
```

```python
print('Shape of matches data',df_players.shape)
```

```python
df_players.info()
```

```python
df_players.isnull().sum()
```

```python
plt.figure(figsize=(15,7))
sns.countplot(x='Bowling_Skill',
              data=df_players,
              order = df_players['Bowling_Skill'].value_counts().index.sort_values()
              )
plt.xlabel('Bowling_Skill',fontsize=12)
plt.ylabel('No of Matches',fontsize=12)
plt.title('Number of matches played in each IPL season',fontsize=16)
```

```python
# Matches won by the Teams
plt.figure(figsize = (11,9))
sns.countplot(y = 'Country',
              data = df_players,
              order= df_players['Country'].value_counts().index,
              palette = 'copper' )
plt.xlabel('No: of Players',fontsize=12)
plt.ylabel('Country',fontsize=12)
plt.title('Number of Players in each Country',fontsize=16)
```
